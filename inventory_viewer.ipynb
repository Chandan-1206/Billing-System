```python
# Step 1: Import required libraries
import tkinter as tk
from tkinter import ttk
import pandas as pd

# Step 2: Load the inventory data
inventory_file = "inventory.csv"  # Ensure this file is in the same directory
inventory_data = pd.read_csv(inventory_file)

# Step 3: Create a function to display the GUI
def display_inventory_gui():
    # Create the main window
    root = tk.Tk()
    root.title("Inventory Viewer")

    # Create a Treeview widget to display the inventory
    tree = ttk.Treeview(root, columns=list(inventory_data.columns), show="headings")
    
    # Define column headings
    for column in inventory_data.columns:
        tree.heading(column, text=column)
        tree.column(column, width=120)
    
    # Insert rows into the Treeview
    for _, row in inventory_data.iterrows():
        tree.insert("", tk.END, values=list(row))
    
    # Pack the Treeview into the window
    tree.pack(expand=True, fill=tk.BOTH)

    # Start the main GUI loop
    root.mainloop()

# Step 4: Call the function to display the GUI
display_inventory_gui()
```

```python
# Step 1: Import required libraries
import tkinter as tk
from tkinter import ttk
import pandas as pd

# Step 2: Load the inventory data
inventory_file = "inventory.csv"
inventory_data = pd.read_csv(inventory_file)

# Step 3: Create a function for the billing system with dynamic dropdown
def billing_system_gui():
    # Create the main window
    root = tk.Tk()
    root.title("Billing System")

    # Frame for item selection
    frame = tk.Frame(root)
    frame.pack(pady=10)

    # Label for item search
    tk.Label(frame, text="Search Item:").grid(row=0, column=0, padx=5, pady=5)

    # Entry widget for searching items
    search_var = tk.StringVar()
    search_entry = tk.Entry(frame, textvariable=search_var)
    search_entry.grid(row=0, column=1, padx=5, pady=5)

    # Listbox for displaying filtered items
    listbox = tk.Listbox(frame, height=5, width=30)
    listbox.grid(row=1, column=0, columnspan=2, padx=5, pady=5)

    # Function to update the listbox based on the search input
    def update_listbox(*args):
        search_text = search_var.get().lower()
        listbox.delete(0, tk.END)  # Clear the listbox
        filtered_items = [item for item in inventory_data["Name"] if search_text in item.lower()]
        for item in filtered_items:
            listbox.insert(tk.END, item)

    # Bind the search variable to update the listbox
    search_var.trace("w", update_listbox)

    # Select item from listbox
    def select_item(event):
        selected_item.set(listbox.get(listbox.curselection()))
        update_price(None)

    listbox.bind("<<ListboxSelect>>", select_item)

    # Variable to store the selected item
    selected_item = tk.StringVar()

    # Radio buttons for Retail/Wholesale
    tk.Label(frame, text="Price Type:").grid(row=2, column=0, padx=5, pady=5)
    price_type = tk.StringVar(value="Retail")
    tk.Radiobutton(frame, text="Retail", variable=price_type, value="Retail").grid(row=2, column=1, padx=5, pady=5, sticky="w")
    tk.Radiobutton(frame, text="Wholesale", variable=price_type, value="Wholesale").grid(row=2, column=2, padx=5, pady=5, sticky="w")

    # Price and Quantity input fields
    tk.Label(frame, text="Price:").grid(row=3, column=0, padx=5, pady=5)
    price_entry = tk.Entry(frame)
    price_entry.grid(row=3, column=1, padx=5, pady=5)

    tk.Label(frame, text="Quantity:").grid(row=4, column=0, padx=5, pady=5)
    quantity_entry = tk.Entry(frame)
    quantity_entry.grid(row=4, column=1, padx=5, pady=5)

    # Function to update price when an item is selected
    def update_price(event):
        item_name = selected_item.get()
        if item_name:
            item_data = inventory_data[inventory_data["Name"] == item_name]
            if not item_data.empty:
                if price_type.get() == "Retail":
                    price_entry.delete(0, tk.END)
                    price_entry.insert(0, item_data["Retail Price"].values[0])
                else:
                    price_entry.delete(0, tk.END)
                    price_entry.insert(0, item_data["Wholesale Price"].values[0])

    # Update price when price type changes
    price_type.trace("w", lambda *args: update_price(None))

    # Calculate total
    def calculate_total():
        try:
            price = float(price_entry.get())
            quantity = int(quantity_entry.get())
            total = price * quantity
            total_label.config(text=f"Total: {total:.2f}")
        except ValueError:
            total_label.config(text="Error: Enter valid price and quantity!")

    # Calculate button
    tk.Button(frame, text="Calculate", command=calculate_total).grid(row=5, column=0, columnspan=2, pady=10)

    # Total label
    total_label = tk.Label(frame, text="Total: 0.00", font=("Arial", 14))
    total_label.grid(row=6, column=0, columnspan=3, pady=5)

    # Run the GUI
    update_listbox()  # Initialize the listbox with all items
    root.mainloop()

# Step 4: Call the billing system GUI
billing_system_gui()
```
